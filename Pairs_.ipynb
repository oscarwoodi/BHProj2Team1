```python
# imports 
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
plt.style.use('fivethirtyeight')
import math
import seaborn as sns
from datetime import datetime
```

```python
# data import 
data = pd.read_excel('Time Series Data.xlsx', index_col = 'Day')

start = 1000
end = 2360

df = data[start:end]
df.head()
```

```python
#Visualise the data

# identified series 25,26,27 as highly correlated
plt.figure(figsize=(20, 10))
plt.plot(df['Series 25']-df['Series 25'][start], linewidth = 1)
plt.plot(df['Series 26']-df['Series 26'][start], linewidth = 1)
plt.plot(df['Series 27']-df['Series 27'][start], linewidth = 1)
plt.title('Stock Adj. Close Price History')
plt.xlabel('Days')
plt.ylabel('Adj. Close Price ($)')
plt.legend(df, loc='upper left')
plt.show()

#Â plotting the difference in prices

a = 'Series 24'
b = 'Series 14'

plt.figure(figsize=(20, 10))
plt.plot(df[b]-df[a]-df[b][start]+df[a][start],\
         linewidth = 1, label='diff')
plt.title('Stock Adj. Diff in Close Price History')
plt.xlabel('Days')
plt.ylabel('Diff ($)')
plt.legend(df, loc='upper left')
plt.show()
```
